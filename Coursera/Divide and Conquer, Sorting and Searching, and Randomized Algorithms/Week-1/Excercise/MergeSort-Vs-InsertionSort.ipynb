```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
n = np.arange(0, 150, 1) #input from 0 ~ 150
```

```python
merge = 6*n * np.log2(n) + 6*n
insert = 0.5*n**2
```

```python
plt.plot(n, merge, n, insert)

plt.legend(('Marge Sort', 'Insert Sort'), loc='upper right')
plt.grid(True)
plt.show()
```
